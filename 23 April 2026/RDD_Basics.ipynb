## **RDD Operations in PySpark**

RDD (Resilient Distributed Dataset) operations are broadly classified into **two categories**:

### **1. Transformations**

Transformations are operations that create a **new RDD** from an existing RDD. These operations are **lazy**, meaning they are not executed immediately. Instead, they are executed only when an action is performed.

**Examples of Transformations:**

* `map()` – Applies a function to each element of the RDD
* `filter()` – Selects elements that satisfy a given condition
* `flatMap()` – Similar to map, but can return multiple elements for each input element
* `distinct()` – Removes duplicate elements from the RDD
* `union()` – Combines two RDDs
* `intersection()` – Returns common elements between two RDDs
* `groupByKey()` – Groups values with the same key
* `reduceByKey()` – Merges values of the same key using a function

---

### **2. Actions**

Actions are operations that **return a result to the driver program** or write data to an external storage system. These operations trigger the execution of transformations.

**Examples of Actions:**

* `collect()` – Returns all elements of the RDD as a list
* `count()` – Returns the number of elements in the RDD
* `first()` – Returns the first element of the RDD
* `take(n)` – Returns the first *n* elements of the RDD
* `reduce()` – Aggregates elements of the RDD using a specified function

```python
rdd=sc.parallelize([1,2,3,4,5])
mapped_rdd=rdd.map(lambda x:x*2)
filtered_rdd=mapped_rdd.filter(lambda x:x>3)
```

```python
filtered_rdd.collect()
```

```python
rdd=sc.parallelize([1,2,3,4,5])
squared_rdd=rdd.map(lambda x:x*x)
squared_rdd.collect()
```

```python
rdd=sc.parallelize(["rahul","sneha","arjun"])
upper_rdd=rdd.map(lambda x:x.upper())
upper_rdd.collect()
```

```python
rdd=sc.parallelize([10,15,20,25,30])
even_rdd=rdd.filter(lambda x:x%2==0)
even_rdd.collect()
```

```python
rdd=sc.parallelize(["apple","banana","mango","kiwi"])
long_words=rdd.filter(lambda x: len(x)>5)
long_words.collect()
```

Flat Map: Can catch multiple elements

```python
lines=sc.parallelize([
    "python spark hadoop",
    "spark sql pysqark"
])

words=lines.flatMap(lambda line:line.split(" "))
words.collect()
```

```python
lines=sc.parallelize(["hello world","pyspark rdd"])
result=lines.map(lambda line:line.split(" "))
result.collect()
```

```python
rdd=sc.parallelize([
    ("Rahul",85),
    ("Sneha",92),
    ("Arjun",78)]
)
rdd.mapValues(lambda x:x+5).collect()
```

```python
rdd=sc.parallelize([10,40,20,30])
rdd.sortBy(lambda x:x).collect()
```

```python
lines=sc.parallelize([
    "python spark pythong",
    "spark hadoop spark",
    "python hadoop"
])

word_counts=(
    lines.flatMap(lambda line: line.split(" "))
    .map(lambda word: (word,1))
    .reduceByKey(lambda a,b:a+b)
)
word_counts.collect()
```

```python
employees=sc.parallelize([
    ("Rahul",70000),
    ("Sneha",60000),
    ("Arjun",75000),
    ("Priya",80000),
    ("Rahul",75000)
])

high_salary=employees.filter(lambda x:x[1]>65000)
high_salary.collect()
```
