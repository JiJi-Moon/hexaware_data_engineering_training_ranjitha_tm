# Section 6 — Joins (Exercises 51–60)

51. Join orders with customers
52. Show customer name and order status
53. Join orders with order_items

55. Create column revenue = quantity * price
56. Join orders, customers, products

59. Calculate revenue per product

````python
orders_customers = orders.join(customers, "customer_id")
orders_customers.select("customer_name","status").show()

order_join = orders.join(order_items,"order_id")
order_items_products = order_items.join(products,"product_id")

revenue_df = order_items_products.withColumn(
    "revenue", col("quantity") * col("product_price")
)

full_join = orders.join(customers,"customer_id") \
    .join(order_items,"order_id") \
    .join(products,"product_id")

full_join.select("customer_name","product_name","quantity").show()

revenue_per_order = revenue_df.groupBy("order_id").agg(sum("revenue")).show()
revenue_per_product = revenue_df.groupBy("product_id").agg(sum("revenue")).show()

revenue_per_customer = full_join.withColumn(
    "revenue", col("quantity")*col("product_price")
).groupBy("customer_id").agg(sum("revenue")).show()
````

````python
from pyspark.sql.functions import sum, col

orders.groupBy("customer_id").count().show()

orders.alias("o").join(customers.alias("c"), "customer_id").groupBy("c.city").count().show()

revenue_df.groupBy("product_id").agg(sum("revenue").alias("total_revenue")).show()

revenue_df.alias("r").join(products.alias("p"), "product_id").groupBy("p.category").agg(sum("r.revenue").alias("total_revenue")).show()

full_join.alias("f").withColumn("revenue", col("quantity") * col("product_price")).groupBy("f.city").agg(sum("revenue").alias("total_revenue")).show()

order_items.groupBy("product_id").agg(sum("quantity").alias("total_quantity")).show()

order_items.alias("oi").join(products.alias("p"), "product_id").groupBy("p.category").agg(sum("oi.quantity").alias("total_quantity")).show()

full_join.alias("f").withColumn("revenue", col("quantity") * col("product_price")).groupBy("f.customer_id").agg(sum("revenue").alias("total_revenue")).show()

full_join.alias("f").withColumn("revenue", col("quantity") * col("product_price")).groupBy("f.customer_id").agg(sum("revenue").alias("total")).orderBy(col("total").desc()).show(5)

revenue_df.groupBy("product_id").agg(sum("revenue").alias("total")).orderBy(col("total").desc()).show(3)
````

````python
logs = sc.textFile("logs.txt")

logs.count()

logs.map(lambda x: x.split(" ")[1]).collect()
logs.map(lambda x: x.split(" ")[0]).collect()

logs.map(lambda x: x.split(" ")[1]).distinct().collect()

logs.filter(lambda x: "login" in x).count()
logs.filter(lambda x: "purchase" in x).count()
logs.filter(lambda x: "view" in x).count()

pairs = logs.map(lambda x: (x.split(" ")[1],1))

pairs.reduceByKey(lambda a,b: a+b).collect()

pairs.reduceByKey(lambda a,b: a+b).sortBy(lambda x: -x[1]).take(1)
````

````python
customers.createOrReplaceTempView("customers")
orders.createOrReplaceTempView("orders")
products.createOrReplaceTempView("products")
payments.createOrReplaceTempView("payments")
order_items.createOrReplaceTempView("order_items")

spark.sql("SELECT * FROM customers").show()

spark.sql("SELECT * FROM customers WHERE age > 30").show()

spark.sql("SELECT city, COUNT(*) FROM customers GROUP BY city").show()

spark.sql("SELECT city, AVG(age) FROM customers GROUP BY city").show()

spark.sql("""
SELECT c.customer_name, o.status
FROM customers c JOIN orders o
ON c.customer_id = o.customer_id
""").show()

spark.sql("""
SELECT customer_id, COUNT(*) FROM orders GROUP BY customer_id
""").show()

spark.sql("""
SELECT p.product_name, oi.quantity
FROM order_items oi JOIN products p
ON oi.product_id = p.product_id
""").show()

spark.sql("""
SELECT oi.product_id, SUM(oi.quantity * p.product_price) AS revenue
FROM order_items oi JOIN products p
ON oi.product_id = p.product_id
GROUP BY oi.product_id
""").show()

spark.sql("""
SELECT c.customer_name, SUM(oi.quantity * p.product_price) AS revenue
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
GROUP BY c.customer_name
ORDER BY revenue DESC
LIMIT 3
""").show()

spark.sql("""
SELECT p.product_name, SUM(oi.quantity * p.product_price) AS revenue
FROM products p
JOIN order_items oi ON p.product_id = oi.product_id
GROUP BY p.product_name
ORDER BY revenue DESC
LIMIT 1
""").show()
````
