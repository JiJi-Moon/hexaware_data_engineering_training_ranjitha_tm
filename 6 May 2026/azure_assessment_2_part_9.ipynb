```python
products_df.write.mode("overwrite").parquet("/tmp/products_parquet_data")
```

```python
para_df=spark.read.parquet("/tmp/products_parquet_data")
para_df.show()
```

```python
para_df.write.format("delta").mode("overwrite").save("/tmp/products_delta_data")
```

```python
spark.sql("""
UPDATE delta.`/tmp/products_delta_data`
SET warehouse_city = 'Mumbai'
WHERE product_id = 113
""").show()
```
