# Part 2 — flatMap Practice

```
sentences = sc.parallelize([
"big data spark",
"spark hadoop hive",
"python spark hadoop",
"spark streaming kafka"
])
```

### Exercise 8

Split all sentences into words
→ `flatMap`

### Exercise 9

Find all distinct technologies
Example:

```
["spark","hadoop","hive","python","kafka"]
```

### Exercise 10

Count occurrences of each word
Expected:

```
spark → 4
hadoop → 2
python → 1
```

Hint pipeline:

```
flatMap → map → reduceByKey
```


````python
sentences = sc.parallelize([
"big data spark",
"spark hadoop hive",
"python spark hadoop",
"spark streaming kafka"
])
````

````python
words=sentences.flatMap(lambda x: x.split(" "))
words.collect()
````

````python
words.distinct().collect()
````

````python
word_map=words.map(lambda x: (x,1))
word_map=word_map.reduceByKey(lambda x,y: x+y)
word_map.collect()
````

# Final Mini Project (RDD)

Using **sales dataset**:

### Tasks:

1. Total revenue per product
2. Top selling product
3. Highest revenue product
4. Products sold more than 3 times
5. Sort products by revenue

### Final Expected Output:

```
Laptop → 375000
Monitor → 48000
Keyboard → 7500
Mouse → 4500
```

````python
revenue=sales_rdd.map(lambda x: (x[0],x[1]*x[2])).reduceByKey(lambda a,b:a+b)
revenue.collect()
````

````python
product_sales=sales_rdd.map(lambda x: (x[0],1)).reduceByKey(lambda a,b:a+b)
product_sales.sortBy(lambda x:x[-1],ascending=False).first()
````

````python
revenue.sortBy(lambda x:x[-1],ascending=False).first()
````

````python
product_sales.filter(lambda x:x[-1]>3).collect()
````

````python
revenue.sortBy(lambda x:x[-1],ascending=False).collect()
````

````python
print("Product Revenue")
for product, revenue in revenue.collect():
    print(f"{product} → {revenue}")
````
