# PART 18 — Applying Functions

Create a custom function that categorizes sales:

> 50000 → Premium  
20000–50000 → Medium  
< 20000 → Small  

Exercise 53  
Apply this function to create sale_category.

---

````python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
def sales_category(amount):
    if amount>50000:
        return "Premium"
    elif amount>20000:
        return "Medium"
    else:
        return "Small"

udf_sales_category=udf(sales_category,StringType())
df_sales.withColumn("sale_category",udf_sales_category(df_sales.amount)).show()
````

# PART 20 — Widgets in Colab

Create widgets to filter employees.

Exercise 56  
Create dropdown for department.

Exercise 57  
Create slider for salary.

Exercise 58  
Display employees based on widget values.

---

````python
# Dropdown for department
import ipywidgets as widgets
from IPython.display import display

dept_dropdown = widgets.Dropdown(
    options=["All"] + [row["department"] for row in df_employees.select("department").distinct().collect() if row["department"] is not None],
    description="Department:"
)

# Slider for salary
salary_slider = widgets.IntSlider(
    value=50000,
    min=0,
    max=100000,
    step=1000,
    description="Min Salary:"
)

display(dept_dropdown, salary_slider)
````

````python
def filter_employees(dept, salary):
    df = df_employees

    if dept != "All":
        df = df.filter(col("department") == dept)

    df = df.filter(col("salary") >= salary)

    df.show()
````

````python
widgets.interactive(
    filter_employees,
    dept=dept_dropdown,
    salary=salary_slider
)
````

# Final Mini Project

Using all datasets, generate report:

Top Employee  
Average Salary  
Department Employee Count  
Total Sales  
Best Sales Employee  

Save results to:

final_report.txt

---

````python
top_employee = df_employees.orderBy(col("salary").desc()).first()
avg_salary = df_employees.select(avg("salary")).first()[0]
dept_count = df_employees.groupBy("department").count().collect()
total_sales = df_sales.select(sum("amount")).first()[0]
best_sales_emp = df_sales.groupBy("emp_id").sum("amount").orderBy(col("sum(amount)").desc()).first()

with open("final_report.txt","w") as f:
    f.write("Top Employee: " + str(top_employee) + "\n")
    f.write("Average Salary: " + str(avg_salary) + "\n")
    f.write("Department Count: " + str(dept_count) + "\n")
    f.write("Total Sales: " + str(total_sales) + "\n")
    f.write("Best Sales Employee: " + str(best_sales_emp))
````
