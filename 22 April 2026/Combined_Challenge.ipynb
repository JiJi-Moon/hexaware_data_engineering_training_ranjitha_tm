## Final Combined Challenge

### Using all datasets:

### Tasks

1. Join `employees.csv` and `sales.csv` using `emp_id`.
2. Find total revenue generated by each employee.
3. Find top 5 employees by sales.
4. Find department generating highest revenue.
5. Save final result to:
   `final_sales_report.csv`

**Example output:**

```
Employee Revenue  
Rahul → 162000  
Priya → 75000  
Arjun → 4500  
```

---

````python
df_joined=df_employees.join(df_sales,"emp_id")
df_joined.show()
````

````python
df_joined.groupBy(df_joined.name).agg(sum(df_joined.revenue).alias("Employee Revenue")).show()
````

````python
df_joined.groupBy(df_joined.name).agg(sum(df_joined.revenue).alias("Employee Revenue")).sort(sum(df_joined.revenue).alias("Employee Revenue").desc()).show(5)
````

````python
df_joined.groupBy(df_joined.department).agg(sum(df_joined.revenue).alias("Department Revenue")).sort(sum(df_joined.revenue).alias("Department Revenue").desc()).show(1)
````

````python
df_emp_revenue=df_joined.groupBy(df_joined.name).agg(sum(df_joined.revenue).alias("Employee Revenue")).sort(sum(df_joined.revenue).alias("Employee Revenue").desc())
````

````python
df_emp_revenue.coalesce(1).write.csv("final_sales_report.csv",header=True)
````
