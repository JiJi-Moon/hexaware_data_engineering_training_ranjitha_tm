## **Section 6 — Joins**

31. Show employees with project budgets by joining employees and projects tables using department.
32. Show employee name, department, project name.
33. Show department-wise project budget.
34. Show employees working in departments that have projects.
35. Count number of employees working per project.

---

````sql
select * from employees e join projects p on e.department=p.department;
````

````sql
select e.emp_name,e.department, p.project_name from employees e join projects p on e.department=p.department;
````

````sql
select e.department,sum(p.budget) ProjectBudget from employees e join projects p on e.department=p.department group by e.department;
````

````sql
select e.* from employees e join projects p on e.department=p.department;
````

````sql
select p.project_id,count(*) EmployeeCount from employees e join projects p on e.department=p.department group by p.project_id;
````

36. Find top 3 highest paid employees in Engineering department.

````sql
select * from employees where department="Engineering" order by salary desc limit 3;
````

````sql
select department from employees group by department having avg(salary) = ( select max(AvgSal) from 
  (select avg(salary) AvgSal from employees group by department)
) ;
````

````sql
select city ,count(*) EmployeeCount from employees group by city having count(*) = (select max(cnt) from 
(select count(*) cnt from employees group by city));
````

````sql
select * from employees where salary >  ( select max(AvgSal) from
(select avg(salary) AvgSal from employees group by department
  )
);
````

## **Final Mini Capstone**

Write SQL queries to create a final analytics table containing:

emp_name
city
department
salary
experience
salary_category
department_avg_salary

---

### **Rules:**

salary_category

salary >= 95000 → High
salary >= 80000 → Medium
else → Low

---

### **Save results:**

```sql
CREATE TABLE employee_salary_analysis AS
SELECT ...
```


````sql
create table employee_salary_analysis as 
select *,case 
when salary>=95000 then "High"
when salary >=80000 then "Medium"
else "Low"
end as salary_category,
avg(salary) over (partition by department) department_avg_salary
from employees;
````

````sql
select * from employee_salary_analysis;
````
