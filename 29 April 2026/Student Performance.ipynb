````python
from pyspark.sql.functions import col, avg, when

data = [
(101, "Arjun Kumar", "Python", 86, 92),
(102, "Neha Gupta", "Python", 74, 80),
(103, "Rahul Nair", "Data Engineering", 88, 85),
(104, "Priya Sharma", "AI", 91, 95),
(105, "Siddharth Rao", "Data Engineering", 67, 72),
(106, "Ayesha Khan", "AI", 79, 88),
(107, "Kavya Reddy", "Python", 94, 96),
(108, "Manish Yadav", "Data Engineering", 58, 69),
(109, "Divya Menon", "AI", 83, 90),
(110, "Farhan Ali", "Python", 62, 75)
]

columns = ["student_id", "student_name", "course", "marks", "attendance"]

df = spark.createDataFrame(data, columns)
display(df)
````

---

### **Tasks**

1. Create a column `grade`:

   * marks ≥ 90 → A
   * marks ≥ 75 → B
   * marks ≥ 60 → C
   * otherwise → Needs Improvement

2. Create a column `certification_status`:

   * marks ≥ 75 and attendance ≥ 80 → Eligible
   * otherwise → Not Eligible

3. Find average marks per course.

4. Count eligible students per course.

---

````python
from pyspark.sql.functions import *
````

````python
df=df.withColumns({"grade":when(df.marks>=90,"A").when(df.marks>=75,"B").when(df.marks>=60,"C").otherwise("Needs Improvement"),"certification_status":when((df.marks>=75)&(df.attendance>=80),"Eligible").otherwise("Not Eligible")})
df.show()
````

````python
df.groupBy(df.course).agg(avg(df.marks).alias("AverageMarks")).show()
````

````python
df.select(df.course,df.certification_status=="Eligible").groupBy(df.course).agg(count("*").alias("Number of Eligible Students")).show()
````
